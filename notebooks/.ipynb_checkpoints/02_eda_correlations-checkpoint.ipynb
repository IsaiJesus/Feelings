## Exploratory Data Analysis

### sleep_level

sleep_level:
- Sin dormir: 0
- Sueño malo: 1
- Sueño medio: 2
- Sueño bueno: 3
- Sueño temprano: 4

mood:
- Horrible: 0
- Mal: 1
- Meh: 2
- Bien: 3
- Increíble: 4

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].scatter(df['sleep_level'], df['mood_numeric'])
axes[0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0].yaxis.set_major_locator(plt.MultipleLocator(1))
axes[0].set_xlabel("sleep_level")
axes[0].set_ylabel("mood")
axes[0].set_title("sleep_level vs mood")

axes[1].scatter(df['sleep_level'], df['emotions'])
axes[1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1].yaxis.set_major_locator(plt.MultipleLocator(1))
axes[1].set_xlabel("sleep_level")
axes[1].set_ylabel("emotions")
axes[1].set_title("sleep_level vs emotions")

axes[2].scatter(df['sleep_level'], df['activities_count'])
axes[2].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[2].yaxis.set_major_locator(plt.MultipleLocator(2))
axes[2].set_xlabel("sleep_level")
axes[2].set_ylabel("activities_count")
axes[2].set_title("sleep_level vs number of activities")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].scatter(df['sleep_level'], df['positive_emotions'])
axes[0].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[0].set_xlabel("sleep_level")
axes[0].set_ylabel("positive_emotions")
axes[0].set_title("sleep_level vs positive_emotions")

axes[1].scatter(df['sleep_level'], df['neutral_emotions'])
axes[1].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[1].set_xlabel("sleep_level")
axes[1].set_ylabel("neutral_emotions")
axes[1].set_title("sleep_level vs neutral_emotions")

axes[2].scatter(df['sleep_level'], df['negative_emotions'])
axes[2].xaxis.set_major_locator(plt.MultipleLocator(1))
axes[2].set_xlabel("sleep_level")
axes[2].set_ylabel("negative_emotions")
axes[2].set_title("sleep_level vs negative_emotions")

plt.tight_layout()
plt.show()
```
