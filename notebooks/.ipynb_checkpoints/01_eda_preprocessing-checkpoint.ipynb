## Data Cleaning

Antes de comenzar el análisis exploratorio de datos, se arreglan los DataFrames, ya que vienen en formatos diferentes, específicamente para el caso de `time` y `date`; además de que no se necesita de las columnas `note_title` ni `note`.

```python
df1 = df1.drop(columns=['date', 'note_title', 'note'])
df2 = df2.drop(columns=['date', 'note_title', 'note'])
```

```python
# Convertir el formato de 'time' en df2 al formato AM - PM
df2['time'] = df2['time'].str.replace(r'\.m\.', 'M', regex=True).str.replace(r'a', 'A', regex=True).str.replace(r'p', 'P', regex=True)
```

```python
# DataFrame del año 2024
df2_filtered = df2[df2['full_date'] <= '2024-12-31']
```

```python
# Por defecto axis=0
df = pd.concat([df2, df1])
```
